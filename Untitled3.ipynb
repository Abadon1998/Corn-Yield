```python
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
df_corn = pd.read_csv('3a Corn Intensification.csv')


df_corn = df_corn.drop(labels=range(28, 56756), axis=0) 

df_corn

```

```python
df_corn_price = (df_corn[['Corn.Price.Lag1']])

df_corn_year = (df_corn[['Year']])

df_corn_yield = (df_corn[['Corn.Yield.(Bu)']])

df_corn_nitrogen = (df_corn[['Nitrogen.Application.Rate']])

df_corn_phosphate = (df_corn[['Phosphate.Application.Rate']])

df_corn_temp = (df_corn[['Temperature.GrowingSeason']])

df_corn_pre = (df_corn[['Precipitation.GrowingSeason']])

df_corn_irr = (df_corn[['Irrigated.Pct']])
```

```python
plt.figure(figsize= (11,4))
a = plt.plot(df_corn_year, df_corn_yield, 'b')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Corn yield over the years')
plt.show(a)

plt.figure(figsize= (11,4))
b = plt.plot(df_corn_year, df_corn_price, 'r')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Corn prices over the years')
plt.show(b)

plt.figure(figsize= (11,4))
c = plt.plot(df_corn_year, df_corn_nitrogen, 'g')
plt.xlabel('Year')
plt.ylabel('Nirtogen Application')
plt.title('Application of Nitrogen')
plt.show(c)

plt.figure(figsize= (11,4))
d = plt.plot(df_corn_year, df_corn_phosphate)
plt.xlabel('Year')
plt.ylabel('Phosphate Application')
plt.title('Application of Phosphate')
plt.show(d)

plt.figure(figsize= (11,4))
e = plt.plot(df_corn_year, df_corn_irr)
plt.xlabel('Year')
plt.ylabel('Phosphate Application')
plt.title('Application of Phosphate')
plt.show(e)
```
